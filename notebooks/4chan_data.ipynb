```
import os
import re
import pandas as pd
from google.colab import drive
drive.mount('/content/drive', force_remount=True)
save_dir = "/content/drive/MyDrive/Classes/Coterm/Winter Quarter/CS224N/CS224N_Project"
data_dir = "/content/drive/Shareddrives/Election Integrity Partnership/Raw Data/4chan/Election Day"
os.chdir(data_dir)
```

```
df = pd.DataFrame()
for num, filename in enumerate(os.listdir(os.getcwd())):
  temp = pd.read_csv(f"{os.getcwd()}/{filename}/election_monitor.csv") 
  df = pd.concat([df, temp], ignore_index=True)
```

```
print(len(df))
unique = df.drop_duplicates()
print(len(unique))
```

```
unique.to_csv(f'{save_dir}/data/4chan_data_raw.csv', index=False) 
```

```
clean = unique[['poster_id', 'comment', 'date/time']]
clean = clean.drop([0,1], axis=0)
clean.head(20)
```

```
clean['comment'] = clean['comment'].apply(lambda x : re.sub(">>[0-9]+", "", str(x))) 
clean['comment'] = clean['comment'].apply(lambda x : re.sub("\n", "", str(x))) 
clean['comment'] = clean['comment'].apply(lambda x : re.sub(">", "", str(x))) 
clean.head()
```

```
clean.to_csv(f'{save_dir}/data/4chan_data.csv', index=False) 
```
