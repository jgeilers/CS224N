```
for i in tqdm(range(166, 63, -1)):
  posts = []
  lines_skip = 0
  num = ''
  if len(str(i)) == 1:
    num = '00' + str(i)
  if len(str(i)) == 2:
    num = '0' + str(i)
  if len(str(i)) == 3:
    num = str(i)
  fl = "data/parler_data/parler_data000000000" + num + ".ndjson"
  with open(fl) as fp:
    for idx, line in enumerate(fp):
      try:
        j_content = json.loads(line)
        post = j_content["body"]
        post = post.replace('\n', ' ')
      except:
        lines_skip += 1
        continue
      if post != "" and post != '\n' and post != " ":
        posts.append(post)
  np_posts = np.array(posts)
  posts = []
  unique_posts = np.unique(np_posts)
  print(unique_posts.shape, lines_skip)
  np.savetxt("data/parler_clean/parler_clean" + str(num) + ".txt", unique_posts,fmt='%s')
```
