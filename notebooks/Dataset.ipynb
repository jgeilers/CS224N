```
from bs4 import BeautifulSoup
import requests
posts = []
header = "https://web.archive.org/web/20210110202718/"
s = ""
with open(txt_files[0],'r') as f:
    s = f.read()
url_list = s.split('\n')

for i in range(20000, len(url_list), 1000):
  j = 0
  for url in tqdm(url_list[i : i + 1000]):
    time.sleep(0.02)
    try:
      a = requests.get(header + url)
      soup = BeautifulSoup(a.content)
      if soup.find('p').findNext('p'):
        posty = soup.find('p').findNext('p').getText()
      else:
        posty = soup.find("p").getText()
      posts.append(posty)
      if j % 50 == 0:
        print(posty)
    except:
      print("Link not working today")
    j += 1
  np_posts = np.array(posts)
  np_posts = np_posts[np_posts != None]
  print(np_posts.shape)
  np.savetxt("parler_posts/parler_last_w_comments_" + str(i) + ".txt", np_posts, delimiter='â€¡',fmt='%s')
```
